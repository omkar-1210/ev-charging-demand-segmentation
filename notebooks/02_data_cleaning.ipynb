```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\loneo\OneDrive\Documents\ev-charging-demand-segmentation\data\raw\EVChargingStationUsage.csv")
```

```python
df = data.copy()
```

```python
cols = [
    'Start Date',
    'End Date',
    'Transaction Date (Pacific Time)',
    'Total Duration (hh:mm:ss)',
    'Charging Time (hh:mm:ss)'
]

for col in cols:
    count = df[col].astype(str).str.match(r'^\d+(\.\d+)?$', na=False).sum()
    print(f"{col}: {count}")
```

```python
mask = df['Transaction Date (Pacific Time)'].astype(str).str.match(r'^\d+(\.\d+)?$', na=False)

df.loc[mask, 'Transaction Date (Pacific Time)'] = pd.to_datetime(pd.to_numeric(df.loc[mask, 'Transaction Date (Pacific Time)']),unit='D',origin='1899-12-30').dt.strftime('%m/%d/%Y %H:%M')

df['Transaction Date (Pacific Time)'] = pd.to_datetime(df['Transaction Date (Pacific Time)'])
```

```python


mask = df['End Date'].astype(str).str.match(r'^\d+(\.\d+)?$', na=False)

df.loc[mask, 'End Date'] = pd.to_datetime(pd.to_numeric(df.loc[mask, 'End Date']),unit='D', origin='1899-12-30').dt.strftime('%m/%d/%Y %H:%M')
```

```python
df['End Date'] = pd.to_datetime(df['End Date'])
```

```python
df['Start Date'] = pd.to_datetime(df['Start Date'])
df['End Date'] = pd.to_datetime(df['End Date'])
df['Transaction Date (Pacific Time)'] = pd.to_datetime(df['Transaction Date (Pacific Time)'])

df['Total Duration (hh:mm:ss)'] = pd.to_timedelta(df['Total Duration (hh:mm:ss)'])
df['Charging Time (hh:mm:ss)'] = pd.to_timedelta(df['Charging Time (hh:mm:ss)'])
```
